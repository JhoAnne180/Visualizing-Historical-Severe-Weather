** load data set**

```python
# Load the severe weather dataset
#data = sns.load_dataset("detail_trim3.csv").sort_values("EVENT_TYPE", ascending=False)
file = pd.read_csv('detail_trim3.csv')

```

```python
count_events_by_month = pd.pivot_table(file,'EVENT_TYPE','MONTH_NAME','STATE', aggfunc= 'count')
count_events_by_month
```

```python
plt.style.use('ggplot')
count_events_by_month.plot.area()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
